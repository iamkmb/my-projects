**THE ADARNA BIRD (Ang Ibong Adarna)**

Once upon a time, in a land far far away, a Kingdom called Berbania existed. It was ruled by King Fernando together with his lovely better half, Queen Valeriana. They were blessed with three sons namely Prince Pedro, Prince Diego, and Prince Juan.

One day,the King became ill. The only cure to the King's illness is the magical Adarna bird, who perches at the tree of Piedras Platas in the mountain of Tabor. The King ordered the 3 princes to catch and bring the bird back to their Kingdom. The succession of the throne will be given as a reward to whoever catches the magical bird.

*Now the adventure begins as the three princes embark their journey to catch the ever elusive Adarna.

```python
# importing random package
import random

# importing time package
import time

    
# defining welcome message function
def welcome_message():
    # Specifying prompt for user to input their name
    user_name = input(prompt = 'Input your name: ').upper()
    # adding time sleep for a little effect
    time.sleep(1)
    
# print a welcome message
    print(f"\nWelcome {user_name} to the Kingdom of Berbania!\n")
    # adding time sleep for a little effect
    time.sleep(1)
    
# print a mission message
    print("""Your mission is to catch the ever elusive Adarna Bird which perches on the tree
called Piedras Platas located at the mountain of Tabor. You have to bring the bird back 
to the Kingdom to heal King Fernando.\n""")
    # adding time sleep for a little effect
    time.sleep(1)
    
# print a warning message
    print("""Just a friendly warning, the Adarna bird can both be a blessing and a curse!
It possesses a magical ability that can heal a deadly illness or turn you into a stone.\n""")
    # adding time sleep for a little effect
    time.sleep(1) 
    
    input(prompt = '<Press ENTER to continue>\n')
    # printing a separator
    print("_*" * 42)

    # calling 1st stage of the game named kingdom
    kingdom()    

# defining the first stage of the game
def kingdom():
    
# creating list of princes
    prince_lst = ["Prince Pedro", "Prince Diego", "Prince Juan"]
    print("""\nTo begin, choose the prince you will accompany in this magical journey.""")
    
# creating user input for choice of prince 
    chosen_prince = input(prompt = """
Choose only from numbers 1 to 3:
           1. Prince Pedro
           2. Prince Diego
           3. Prince Juan \n""")
    # while loop to set up a selection of princes
    while True:
        # try and except was placed to catch input errors
        try:
            # converting the user input to int
            chosen_prince = int(chosen_prince)
        except:
        # printing message if input was wrong or out of range
            print("""\nNo prince from Berbania was selected. This means, you opted to embark
on this quest alone. Brave fella!\n""")
            break
            
        # Check if the input is within the valid range
        if chosen_prince and 1 <= int(chosen_prince) <= 3:
        # Get the corresponding prince
            chosen_prince = prince_lst[int(chosen_prince) - 1]
            print(f"You chose to accompany: {chosen_prince}")
            break
            
        else:
            print("""\nNo prince from Berbania was selected. This means, you opted to embark 
on this quest alone. Brave fella!\n""")
            break           
            
    input(prompt = '<Press ENTER to continue>\n')
    
    # this moves the user to the enchanted forest
    enchanted_forest()
# adding stage 2
def enchanted_forest():
    # printing a separator
    print("_*" * 42)
    
    print("""\nThe gatekeeper greeted you to the enchanted forest. He will open the gate
for you and lead you the hermit's home. The hermit possesses a box which contains items 
that will help you catch the Adarna bird and avoid turning into a stone.

But first, you have to outwit the gatekeeper by answering his random old-fashioned
riddle, impress him through your singing skill, or get him groove with you through 
your dancing ability.\n""")

    input("""<Press ENTER when you are ready. You only have 2 attempts to impress 
the gatekeeper.>\n""")
    # Set the number of attempts
    attempts = 2

    # setting up a for loop
    for attempt in range(1, attempts + 1):
        print(f"\nAttempt {attempt} of {attempts}")
        user_choice = input(prompt = """Choose which one you want to do, 'riddle', 'sing', 'dance': 
>""").lower()

        # conditional statement if user chooses dance
        if user_choice == 'dance':
            print("You gave it your all dancing to the tune of Dancing Queen.\n")
            input(prompt = "Hit ENTER if you want to know the gate keepers reaction.\n")
            if random.choice([True, False]):
                print("""Great job! The gate keeper was impressed, you may pass 
the gate of enchated forest. The gatekeeper will lead you to the hermit's home!\n""")
                break
                mount_tabor()
            else:
                print("Oh no! You're such a horrible dancer.\n")
        
        # conditional statement if user chooses sing        
        elif user_choice == 'sing':
            print("You sang your heart out...")
            input(prompt = """Press ENTER if you want to know the gate keepers reaction to 
your rendition of Taylor Swift's 10-minute version of All too well?\n""")
            if random.choice([True, False]):
                print("""You're quite a singer right there. You may pass the gate
of enchated forest. The gatekeeper will lead you to the hermit's home!\n""")
                break
                mount_tabor()
            else:
                print("""Oops! I may not know that song but one thing is for sure,
you're out of tune. The gatekeeper doesn't want what he is hearing!\n""")
        
        # conditional statement if user chooses riddle        
        elif user_choice == 'riddle':
            print("You answered the riddle of the gate keeper with confidence.\n")
            input(prompt = "Press ENTER for the result.")
            if random.choice([True, False]):
                print("""\nYou're impressive! You were able to answer such an old 
riddle. You may pass the gate of enchated forest. The gatekeeper will lead you to
the hermit's home!\n""")
                break
                mount_tabor()
            else:
                print("Sorry, but that's not the answer to the riddle.\n")
                
        else: 
            print("""\nYou didn't seem to impress the gatekeeper. Will he let you pass by?\n""")
            input(prompt = "Press ENTER if you want to know the answer.\n")
            
    # print statement if 2 attempts are exhausted or 2 wrong input was given
    else:
        print("""He did let you pass through, knowing you have another challenge waiting ahead of you.\n""")
    
    input(prompt = '<Press ENTER to continue>\n')
    
    # calling the next stage
    mount_tabor()      
def mount_tabor():
    # to put a separator between stages
    print("_*" * 42)
    
    print("""\nYou have arrived at the hermit's humble abode! The hermit was known 
to be generous, he offers a mysterious box which he dub as "a man's survival kit
to catching the Adarna bird".\n""")
    

    ## declaring objects
    number = random.randint(1,5) 
    guesses = 2

# using try and except
# looping until a user is out of guesses
    while guesses > 0:
        choice2 = input (prompt = """\nGive a random number from 1-5. If you guess the number which
the hermit wrote on the box, the mysterious box will be yours.\n>""")
    
        try:

        # converting choice for numerical comparison
            choice2 = int(choice2)
        except:
            print("""That's a good try, but I need a number. Remember, your fate
relies on the box.""")
    
        # conditional statement based on user input
        # guessed correctly,will give more information about the bird + the box
        if choice2 == number:
            print ("""Great! You have guessed the number correctly. 
The Adarna bird will sing 7 songs, after the 7th and last song, it defecates and 
sleeps with eyes wide-open. Once it sleeps, it's the time you'll throw the golden 
rope over the Adarna and catch it. 

But before that, once the bird started singing you have cut the palm of your hand 
using the knife and squeeze a lime into it to fight the drowsiness brought by every 
song.""")
            win()
            break
            input(prompt = '<Press any key to continue>\n')
        # incorrect answer    
        elif choice2 != number:
            guesses -= 1
        
            if guesses > 0:
                print(f"\nNice try! You have {guesses} guesses left.")
        
            elif guesses == 0:
                print(f"\nYou run out of guesses my friend!")
                input(prompt = '<Press any key to continue>\n')
                fail()
                
        # setting a different error message for inner and outer loops
        else:
            print("\nYou didn't guessed correctly. I'm afraid you're on your own.")
            input(prompt = '<Press any key to continue>\n')
            fail()
            

## writing a win function
def win():
    print("""\nThe Kingdom is rejoicing on your triumph! You will be rewarded handsomely.\n""")
## writing a fail function    
def fail():
    print("""\nYou failed King Fernndo on this mission. The Kingdom is lamenting.""")


# Calling the first defined function, this will start the game        
welcome_message()
```
